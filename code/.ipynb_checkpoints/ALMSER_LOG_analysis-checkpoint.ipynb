# Plot results from one log file

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import collections
import ast



path = "/home/aprimpel/jupyter_notebooks/ICML2021_GBALMSER/datasets/magellan_restaurants/almser/pick_conflict_0_5/"
log_file = pd.read_csv(path+"1_runs_200_iter_disagreement_post_graph_24_02_01_29___log.csv")


print("Evaluate bridges and minimum cuts removal:")
print("---")
print("In total removed %i bridges of which %i were wrong (%.2f)" %(log_file.bridges.sum(),log_file.wrong_bridges.sum(),log_file.wrong_bridges.sum()/log_file.bridges.sum()))
print("In total removed %i cuts of which %i were wrong (%.2f)" %(log_file.cuts.sum(),log_file.wrong_cuts.sum(),log_file.wrong_cuts.sum()/log_file.cuts.sum()))
print("---")
print("***************************************************************************************")

print("Evaluate the correctness of the additional training data per iteration:")
print("---")
fig, ax = plt.subplots()
train_acc, = ax.plot(log_file['addit_train_data_acc'], color="#E69F00", label="Additional Train Data Accuracy")
train_f1, = ax.plot(log_file['addit_train_data_f1'], color= "#56B4E9", label="Additional Train Data F1")
ax.grid(True)


plt.xticks(fontsize=15)
plt.yticks(np.arange(0.0, 1.1, step=0.02), fontsize=15)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.1),handles=[train_acc,train_f1], fontsize=12)
plt.axis([0,200,0.8,1.0])
plt.show()
print("---")

print("***************************************************************************************")

print("Compare correctness of additional training data to the complete graph")
print("---")
fig, ax = plt.subplots()
train_acc, = ax.plot(log_file['addit_train_data_acc'], color="#E69F00", label="Additional Train Data Accuracy")
graph_acc, = ax.plot(log_file['all_graph_acc'], color="#009E73", label="Complete Graph Accuracy")
ax.grid(True)

plt.xticks(fontsize=15)
plt.yticks(np.arange(0.0, 1.1, step=0.01), fontsize=15)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.1),handles=[train_acc,graph_acc], fontsize=12)
plt.axis([0,200,0.95,1.0])
plt.show()
print("---")

print("***************************************************************************************")

print("Compare correctness (F1) of additional training data to the complete graph")
print("---")
fig, ax = plt.subplots()
train_f1, = ax.plot(log_file['addit_train_data_f1'], color= "#56B4E9", label="Additional Train Data F1")
graph_f1, = ax.plot(log_file['all_graph_f1'], color= "#0072B2", label="Complete Graph F1")
ax.grid(True)

plt.xticks(fontsize=15)
plt.yticks(np.arange(0.0, 1.1, step=0.02), fontsize=15)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.1),handles=[train_f1, graph_f1], fontsize=12)
plt.axis([0,200,0.8,1.0])
plt.show()
print("---")

print("***************************************************************************************")

print("Size of additional training data per iteration")
print("---")
fig, ax = plt.subplots()
train_data_size, = ax.plot(log_file['addit_train_data_size'], color="#D55E00", label="# Additional Train Record Pairs per Iteration")
ax.grid(True)
plt.show()
print("---")

print("***************************************************************************************")
print("Small VS Large Connected Components per iteration")
print("---")
fig, ax = plt.subplots()
small_cc, = ax.plot(log_file['small_cc'], color="#CC79A7", label="# Small CC per Iteration")
large_cc, = ax.plot(log_file['large_cc'], color="#F0E442", label="# Large CC per Iteration")
ax.grid(True)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.1),handles=[small_cc, large_cc], fontsize=12)
plt.show()
print("---")

print("***************************************************************************************")
print("Connected Component size distribution in different snapshots")
print("---")
snaps = [21,99,199]
for snap in snaps: 

    distr_= ast.literal_eval(log_file['cc_distribution'].loc[snap].replace("Counter(","").replace(")",""))
    ordered_dist = collections.OrderedDict(sorted(distr_.items()))
    
    keys_ = []
    values_ = []
    for key, value in ordered_dist.items():
        keys_.append(key)
        values_.append(value)
    
    print("Largest Connected Component is of size:", max(keys_))
    plt.bar(keys_,values_)
    plt.title("Connected Component size distribution at %i iteration" %(snap+1))
    plt.show()
print("---")
print("***************************************************************************************")
```

# Plot F1 scores from multiple files

```python
path = "/home/aprimpel/jupyter_notebooks/ICML2021_GBALMSER/datasets/magellan_restaurants/almser/"
log_file_no_graph_qs = pd.read_csv(path+"pick_conflict_0_0/1_runs_200_iter_disagreement_post_graph_23_02_03_42.csv")
log_file_almser_gb = pd.read_csv(path+"pick_conflict_0_5/1_runs_200_iter_disagreement_post_graph_24_02_00_32.csv")
log_file_full_graph_qs = pd.read_csv(path+"pick_conflict_1_0/1_runs_200_iter_disagreement_post_graph_23_02_16_08.csv")


print("Compare F1 scores on the test set")
print("---")
fig, ax = plt.subplots()
f1_no_graph, = ax.plot(log_file_no_graph_qs['F1_micro'], color="#E69F00", label="QS with 0% graph signal")
f1_half_graph, = ax.plot(log_file_almser_gb['F1_micro'], color="#009E73", label="QS with 50% graph signal")
f1_full_graph, = ax.plot(log_file_full_graph_qs['F1_micro'], color="#0072B2", label="QS with 100% graph signal")

ax.grid(True)

plt.xticks(fontsize=15)
plt.yticks(np.arange(0.0, 1.1, step=0.05), fontsize=15)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.02),handles=[f1_no_graph,f1_half_graph,f1_full_graph], fontsize=12)
plt.axis([0,200,0.8,1.0])
plt.show()
print("---")
print("***************************************************************************************")

```
